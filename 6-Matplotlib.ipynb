# MatplotLib
MatplotLib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt or GTK+.

Some of the major Matplotlib pros are: 

- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general
- 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
## Simple Examples
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
## plotting using matplotlib
## plot Scatter

plt.scatter(x,y, c='y')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Graph in 20')
plt.savefig('Test.png')
```

```python
#plt plot
plt.plot(x,y, 'bo')
```

```python
y=x*x
```

```python
plt.plot(x,y, 'y*', linestyle='dashed', linewidth=3, markersize=12)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('2D Diagram')
```

```python
#Creating Subplots
plt.subplot(2,2,1)
plt.plot(x,y, 'r--', markersize=10)
plt.subplot(2,2,2)
plt.plot(x,y, 'y*-', markersize=10)
plt.subplot(2,2,3)
plt.plot(x,y, 'go-', markersize=10)
```

```python
x=np.arange(1,11)
y=3*x+5

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('2D Diagram')
plt.plot(x,y)
plt.show()
```

```python
# compute the x and y coordinates for points on a sine curve

x=np.arange(0,4 * np.pi, 0.1)
y=np.sin(x)
plt.title('sine wave form')

# plot the points using the matplotlib
plt.plot(x,y)
plt.show()
```

```python
# Compute the x and y coordinates for points on sine and cosine curves
x=np.arange(0,5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Setup a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt.subplot(2,1,1)

# MAke the first plot
plt.plot(x,y_sin, 'r--')
plt.title('Sine')
plt.show()
# Make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos, 'g*-')
plt.title('cosine')
plt.show()
```

# Bar Graph

```python
## Bar plot

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y, align= 'center')
plt.bar(x2,y2, color= 'g')
plt.title('Bar Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

# Histograms

```python
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,29])
plt.hist(a)
plt.title('Histogram')
plt.show()
```

# Box plot using Matplotlib

```python
data=[np.random.normal(0, std, 100) for std in range (1,4)]

# Rectangular Box Plot
plt.boxplot(data,vert=True,patch_artist=True);
```

# Pie Chart

```python
# Data to plot

labels= 'Python', 'Ruby', 'C++', 'Java'
sizes= [215,130,245,210]
colors=['yellowgreen', 'gold', 'lightcoral', 'lightskyblue']
explode=(0.1, 0,0,0) #explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
